```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Sample data
data = pd.DataFrame({
    'Category': ['A', 'B', 'C'],
    'Value': [10, 24, 36]
})

# 1. Matplotlib
plt.bar(data['Category'], data['Value'], color='skyblue')
plt.title('Matplotlib Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 2. Seaborn
sns.barplot(x='Category', y='Value', data=data, palette='viridis')
plt.title('Seaborn Bar Chart')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 3. Plotly
fig = px.bar(data, x='Category', y='Value', title='Plotly Bar Chart')
fig.show()
```
