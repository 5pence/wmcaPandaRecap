```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
df_penguins = sns.load_dataset("penguins")

# Quick look
df_penguins.head()
```

```python
# drop non-numeric columns
df_num = df_penguins.select_dtypes(include=["float64", "int64"]).dropna()

# show columns
df_num.columns
```

```python
summary_penguins = pd.DataFrame({
    "Mean": df_num.mean(),
    "Median": df_num.median(),
    "Mode": df_num.mode().iloc[0],
    "Std Dev": df_num.std(),
    "Variance": df_num.var(),
    "Skewness": df_num.skew(),
    "Kurtosis": df_num.kurtosis(),
    "Min": df_num.min(),
    "Max": df_num.max(),
    "Range": df_num.max() - df_num.min(),
})

summary_penguins.style.background_gradient(cmap="Greens").format(precision=2)
```

```python
col = "body_mass_g"

fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(df_num[col], kde=True, bins=30, color="skyblue")
plt.axvline(df_num[col].mean(), color='r', linestyle="--", label="Mean")
plt.axvline(df_num[col].median(), color='k', linestyle="--", label="Median")
plt.axvline(df_num[col].mode()[0], color='g', linestyle="--", label="Mode")
plt.title(f"{col} Distribution with Mean/Median/Mode")
plt.legend()
plt.show()
```
